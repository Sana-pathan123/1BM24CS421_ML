```python
import math

# Training data (x, y, class)
data = [(1,2,'A'), (2,3,'A'), (3,3,'B'), (6,5,'B'), (7,7,'B')]
test = (3,4)
K = 3

print("=== KNN CLASSIFICATION ===")

# Step 1: Distance calculation
distances = []
print("\nStep 1: Distances")
for x,y,label in data:
    d = math.sqrt((x-test[0])**2 + (y-test[1])**2)
    distances.append((d,label))
    print(f"({x},{y}) → {round(d,2)}")

# Step 2: Sort
distances.sort()
print("\nStep 2: Sorted")
for d in distances:
    print((round(d[0],2), d[1]))

# Step 3: K nearest
k_near = distances[:K]
print(f"\nStep 3: {K} Nearest")
for d in k_near:
    print((round(d[0],2), d[1]))

# Step 4: Voting
votes = {}
for _,label in k_near:
    votes[label] = votes.get(label,0) + 1

prediction = max(votes, key=votes.get)
print("\nFinal Prediction:", prediction)
```

```python
import math

# Training data (x, y, value)
data = [(1,2,10), (2,3,12), (3,3,14), (6,5,20), (7,7,25)]
test = (3,4)
K = 3

print("\n=== KNN REGRESSION ===")

# Step 1: Distance calculation
distances = []
print("\nStep 1: Distances")
for x,y,value in data:
    d = math.sqrt((x-test[0])**2 + (y-test[1])**2)
    distances.append((d,value))
    print(f"({x},{y}) → {round(d,2)}")

# Step 2: Sort
distances.sort()
print("\nStep 2: Sorted")
for d in distances:
    print((round(d[0],2), d[1]))

# Step 3: K nearest
k_near = distances[:K]
print(f"\nStep 3: {K} Nearest")
values = [v for _,v in k_near]
print(values)

# Step 4: Average
prediction = sum(values)/K
print("\nFinal Prediction:", prediction)
```
